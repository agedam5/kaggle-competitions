```python
import numpy as np
import pandas as pd
import random
```

```python
t_train = pd.read_csv("train.csv")
t_test = pd.read_csv("test.csv")
```

```python
t_train = t_train.drop(['PassengerId','Name','Ticket','Cabin'],axis=1)
t_test = t_test.drop(['PassengerId','Name','Ticket','Cabin'],axis=1)
```

```python
t_train['Age'] = t_train['Age'].fillna(value = t_train['Age'].mean())
t_test['Age'] = t_test['Age'].fillna(value = t_test['Age'].mean())
t_test['Fare'] = t_test['Fare'].fillna(value = t_test['Fare'].mean()) 
```

```python
t_test.isnull().sum()
```

```python
t_train['Survived'].value_counts()
```

```python
t_train['Embarked'] = t_train['Embarked'].fillna(value=t_train['Embarked'].mode().loc[0])
```

```python
t_train.isnull().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
t_train = pd.get_dummies(t_train,drop_first=True)
t_test = pd.get_dummies(t_test,drop_first=True)
```

```python
t_train.head()
```

```python
y_train = t_train.iloc[:,0:1]
t_train = t_train.drop(['Survived'],axis=1)
```

```python
scaler = MinMaxScaler()
t_train[['Age','Fare']] = scaler.fit_transform(t_train[['Age','Fare']])
t_test[['Age','Fare']] = scaler.fit_transform(t_test[['Age','Fare']])
```

```python
t_train.head()
```

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(t_train,y_train)
```

```python
y_pred = pd.DataFrame(clf.predict(t_test))
```

```python
y_pred
```

```python
submission = pd.read_csv("test.csv")
```

```python
submission=submission.drop(['Pclass','Name','Sex','Age','SibSp','Parch','Ticket','Fare','Cabin','Embarked'],axis=1)
```

```python
submission = submission.join(y_pred)
```
